```python
import networkx as nx

G = nx.DiGraph() #무슨 빈그래프를 생성한다. Di는 Directed graph라고 이해하시면될듯하다 
G.add_nodes_from(['A', 'B', 'C', 'D', 'E']) # 노드를 5개 추가한다. 
G.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D')]) 
#add_edges_from은 선을 어떻게 연결하는가를 표현해준다 A는 B와  연결 B는 c와 연결 c는 d와 연결해준다.


nx.draw(G, with_labels=True)
# 그림을 보면 a에서 b로 b에서 c로 c에서 d로 가는것을 알수있다. 
```

노드 추가및 제거하기와 선 추가및 제거하기에 대해 알아보자 

```python
#노드 추가 및제거하기
g1 = nx.Graph()
g1.add_node('David')
g1.add_node('John')
g1.add_node('Mary')
g1.add_node('Julia')
g1.add_node('Juda')
g1.add_nodes_from(['Judy', 'Karen'])
g1.remove_node('Julia')
 #선 추가 및제거하기
g1.add_edge('David', 'John')
g1.add_edge('David', 'Mary')
g1.add_edge('David', 'Karen')
g1.add_edges_from([('John', 'Mary'), ('John', 'Judy')])
g1.remove_edge('Mary', 'John')
nx.draw(g1, with_labels=True, font_weight='bold') #글꼴은 bold체이고 label은 다 이름을 써줘!
```

노드 크기를 차수에 비례하여 설정하기 

```python
d = dict(g1.degree) 
nx.draw(g1, nodelist= d.keys(), node_size = [v*1000 for v in d.values()], \
# for 반복문을 통해서 , 크기를조절하는듯하다. 
```

```python
 #연관규칙데이터프레임생성하기

result = (list(apriori(dataset, min_support=0.01))) #apriori함수를 이용해 불러내서 result에 할당..하는듯, 최소지지도가 0.01인것
df= pd.DataFrame(result)

#그 중에서 항목수가2인것만남기기
df['length'] = df['items'].apply(lambda x: len(x))
df= df[(df['length']==2)&(df['support']>=0.01)].sort_values(by='support',ascending=False)

df.head(3)
                                                            df.head(3)
```
